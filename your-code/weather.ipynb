### Importing the data

```python

import pandas as pd
weather = pd.read_csv('C:/Users/victo/Desktop/IronHack/13-lab-pandas-data-cleaning/your-code/weather-raw.csv')
weather
```

### Using melt for moving d1 to d30 into a single column. Also changing the column name to days

```python
weather1 = weather.melt(id_vars=['id', 'year', 'month', 'element'], value_vars=['d1', 'd2', 'd3', 'd4', 'd5', 'd6', 'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13', 'd14', 'd15', 'd16', 'd17', 'd18', 'd19', 'd20', 'd21', 'd22', 'd23', 'd24', 'd25', 'd26', 'd27', 'd28', 'd29', 'd30', 'd31'], var_name='days')
weather1
```

###  We drop NA

```python
weather2=weather1.dropna(axis=0, how='any')
weather2
```

###  We use pivot_table for having the summary of tmax and tmin of each month in 2010

```python
weather3 = pd.pivot_table(weather2,index='month', columns = 'element', values='value')
weather3
```
